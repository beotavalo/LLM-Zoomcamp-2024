```python
from sentence_transformers import SentenceTransformer
model_name = 'multi-qa-mpnet-base-dot-v1'
embedding_model = SentenceTransformer(model_name)
```

```python
answer_llm = df.iloc[0].answer_llm
answer_llm
```

```python
answer_embedding = embedding_model.encode(answer_llm)
answer_embedding
```

```python
# Print the first value of the resulting vector
first_value = answer_embedding[0]
print(f"First value: {first_value}")
```

```python
all_llm_answers = df.answer_llm #Select all LLM answers
all_orig_answers = df.answer_orig #Select all original answers
v_llm = embedding_model.encode(all_llm_answers)
v_orig = embedding_model.encode(all_orig_answers)
```

```python
v_llm[0], v_orig[0]
```

```python
import tqdm

vector_df = pd.DataFrame()
vector_list = []

for i in range (len(v_llm)):
    dot_prod = v_llm[i].dot(v_orig[i])
    vector_list.append(dot_prod)

vector_df['result'] = vector_list
```

```python
# Normalize in this way [From Q2, we can see that the results are not within the [0, 1] range. It's because the vectors coming from this model are not normalized.
# So we need to normalize them.
# To do it, we
#     Compute the norm of a vector
#     Divide each element by this norm
# So, for vector v, it'll be v / ||v||]

import numpy as np

v_llm_norm = np.linalg.norm(v_llm, axis=1, keepdims=True)
v_orig_norm = np.linalg.norm(v_orig, axis=1, keepdims=True)
print(v_llm_norm[0], v_orig_norm[0])
v_llm_normalized = v_llm / v_llm_norm
v_orig_normalized = v_orig / v_orig_norm

vector_list_normalized = []
for i in range(len(v_llm_normalized)):
    dot_prod_normalized = v_llm_normalized[i].dot(v_orig_normalized[i])
    vector_list_normalized.append(dot_prod_normalized)

vector_df_normalized = pd.DataFrame()
vector_df_normalized['result'] = vector_list_normalized
vector_df_normalized.describe()
```

```python
from rouge import Rouge
rouge_scorer = Rouge()
```

```python
scores = rouge_scorer.get_scores(df['answer_llm'], df['answer_orig'])
```

```python
# Get the score of index 10
scores[11]
```

```python
# Get 'r' form the scores[11] for rouge-1, rouge-2, rouge-l

r1 = scores[11]['rouge-1']['f']
r2 = scores[11]['rouge-2']['f']
rl = scores[11]['rouge-l']['f']

average_rouge = (r1 + r2 + rl) / 3
average_rouge

```

```python
# Let's compute the score for all the records
# rouge_1 = scores['rouge-1']['f']
# rouge_2 = scores['rouge-2']['f']
# rouge_l = scores['rouge-l']['f']
# rouge_avg = (rouge_1 + rouge_2 + rouge_l) / 3
# And create a dataframe from them
# What's the agerage rouge_2 across all the records?

rouge_1 = []
rouge_2 = []
rouge_l = []
rouge_avg = []

for score in scores:
  r1 = score['rouge-1']['f']
  r2 = score['rouge-2']['f']
  rl = score['rouge-l']['f']
  avg = (r1 + r2 + rl) / 3

  rouge_1.append(r1)
  rouge_2.append(r2)
  rouge_l.append(rl)
  rouge_avg.append(avg)

rouge_df = pd.DataFrame({
    'rouge_1': rouge_1,
    'rouge_2': rouge_2,
    'rouge_l': rouge_l,
    'rouge_avg': rouge_avg
})

```
